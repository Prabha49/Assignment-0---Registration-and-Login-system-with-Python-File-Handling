```python
import re
import csv
```

## CSV Imported for handling csv files

##Register() function is used for Register new users

```python
def Register():
  print("NOTE: "+"\n"+"Your Password Must Be Contain Minimum One Special Character, One Digit, One Uppercase, One Lowercase Character ")
  print("Password Lenght Must Be in (5 < password length > 16)")
  Name = input("Enter Your Name: ")
  Mail_ID = input("Enter Mail Id: ")
  Pws = input("Enter Your Password: ")
  IStatus = Verify_ID(Mail_ID)               #Checking The Entered Mail Id is In Correct Format Or Not
  if IStatus == True:                        #If Yes
    MStatus = Check_Exits(Mail_ID)           #Checking The Entered Mail Id is Already Exits Or Not
    if MStatus == True:
      print("The Mail Id Already Exits")
  elif IStatus == False:
    print("Invalid Mail Id")
    print("Enter Correct Format Mail Id")
  PStatus = Password(Pws)                    #The Mail Not Exits Then it Will Check The Password Is in Correct Format Or Not
  if PStatus == True and MStatus == False:   #if Password Is in Correct Format Then All Credentials Will Be Updated In LogData.csv File 
    Update(Name, Mail_ID, Pws)
    print("You Are Registed")
  elif PStatus == False:
    print("Your Password is Not in Correct Format")
    print("Enter Correct Format Password")
```

##Verify_ID() Function Used For checking the user entered mail id is in correct format or not

```python
def Verify_ID(Mail_ID):
  pattern = '^[a-z0-9\.\_]+@[a-z0-9]+\.[com | org | in]+$'
  Mail_Verify = re.match(pattern,Mail_ID)
  if Mail_Verify is None : 
      return False
  else:
      return True
```

##Password() Function Used For checking the user entered password is in valid format or not

```python
def Password(Pws):
	l, u, sp, d = 0, 0, 0, 0
	specialchar="!@#$%^&"
	if (len(Pws)>=5 & len(Pws) <= 16):
		for i in Pws:
			# counting lowercase alphabets
			if (i.islower()):
				l+=1		
			# counting uppercase alphabets
			if (i.isupper()):
				u+=1		
			# counting digits
			if (i.isdigit()):
				d+=1		
			# counting the mentioned special characters
			if (i in specialchar):
				sp+=1		
	if (l>=1 and u>=1 and sp>=1 and d>=1 and l+sp+u+d==len(Pws)):
		#print("Valid Password")
 		return True
	else:
		#print("Invalid Password")
	  return False
```

##Update() Function Is Used For Uploding The Valid User Details In CSV File

```python
def Update(Name, Mail_ID, Pws):
  Log_Details = {'Name':Name, 'Mail_ID':Mail_ID,'Password':Pws }
  Field_Name = ["Name", "Mail_ID", "Password"]
  File_Name = 'LogData.csv'
  with open (File_Name, 'a+') as New_CSV:
    Writer = csv.DictWriter(New_CSV, fieldnames=Field_Name, delimiter = "," )
    Writer.writerow(Log_Details)
```

##LogAccess() is used for Login access

```python
def LogAccess():
  print("Enter Your Credentials")
  Name = input("Enter Your Name: ")
  Mail_ID = input("Enter Mail Id: ")
  Pws = input("Enter Your Password: ")
  LStatus = Verify_all(Name, Mail_ID, Pws)
  if LStatus == True:
    print("Welcome")
    print("You Are Logged")
  else:
    print("This Mail Id Not Yet Registered, Try To Register")
```

##Verify_all()function get all credential deails form LogAccess() and verify it presented in csv file or Not

```python
def Verify_all(Name, Mail_ID, Pws):
  Log_Keys = ["Name", "Mail_ID", "Password"]
  with open ('LogData.csv', 'r') as New_CSV:
    Reader = csv.DictReader(New_CSV)
    Collection = { col_name: [] for col_name in Log_Keys }
    for row in Reader:
      for col_name in Log_Keys:
        Collection[col_name].append(row[col_name])
    for j in range(len(Collection['Mail_ID'])):
      if Collection['Name'][j] == Name and Collection['Mail_ID'][j] == Mail_ID and Collection['Password'][j] == Pws:
        return True
```

##ForPass() is used for Retrive the Forgetted original password For user

```python
def ForPass():
  Name = input("Enter Your Name: ")
  Mail_ID = input("Enter Mail Id: ")
  Log_Keys = ["Name", "Mail_ID", "Password"]
  with open ('LogData.csv', 'r') as New_CSV:
    Reader = csv.DictReader(New_CSV)
    Collection = { col_name: [] for col_name in Log_Keys }
    for row in Reader:
      for col_name in Log_Keys:
        Collection[col_name].append(row[col_name])
      for j in range(len(Collection['Mail_ID'])):
        if Collection['Name'][j] == Name and Collection['Mail_ID'][j] == Mail_ID:
          print("Your Password is ", Collection['Password'][j])
          return True
        else:
          return False
```

## Check_Exits() is used for checking the mail id is already presented or not in csv file

```python
def Check_Exits(Mail_ID):
  Log_Keys = ["Name", "Mail_ID", "Password"]
  Mail_ID = Mail_ID
  with open ('LogData.csv', 'r') as New_CSV:
    Reader = csv.DictReader(New_CSV)
    Collection = { col_name: [] for col_name in Log_Keys }
    for row in Reader:
      for col_name in Log_Keys:
        Collection[col_name].append(row[col_name])
    for j in range(len(Collection['Mail_ID'])):
      if Collection['Mail_ID'][j] == Mail_ID:
        return True
      else:
        return False
```

##This block of code is the main program

```python
Close = 1
print("Welcome, please select an option")

while Close:
  option = input("Login | Register | Forgot Password | Close ")
  if option == "Login":
    LogAccess()
  elif option == "Register":
    Register()
  elif option == "Forgot Password":
    FStatus = ForPass()
    if FStatus == False:
      print("Invalid Credentials")
      print("Try To Register")
  elif option == "Close":
    Close = 0
  else:
    print("Please enter a valid parameter, this is case-sensitive")
```
